###  1- Doughnut chart having rings for cut type amd each ring shows part to whole relationship of color for the respective cut type. 

```python
doughnut = pd.DataFrame(diamonds.groupby(['cut', 'color']).size()).reset_index().rename(columns= {0:'count'})
sns.set(font_scale=1.2)
plt.figure(figsize=(15,15))

pie = plt.pie(doughnut[doughnut['cut']=='Fair']['count'],labeldistance = 1.1,autopct = '%0.01f%%',pctdistance=0.95,startangle = 90,
            wedgeprops = dict(width = 0.5),radius = 4,explode=(0.01,0.01,0.01,0.01,0.01,0.01,0.01))

pie = plt.pie(doughnut[doughnut['cut']=='Good']['count'],labeldistance = 1.1,autopct = '%0.01f%%',pctdistance=0.95,startangle = 90,
            wedgeprops = dict(width = 0.5),radius = 3.45,explode=(0.01,0.01,0.01,0.01,0.01,0.01,0.01),)

pie = plt.pie(doughnut[doughnut['cut']=='Ideal']['count'],labeldistance = 1.1,autopct = '%0.01f%%',pctdistance=0.9,startangle = 90,
            wedgeprops = dict(width = 0.5),radius = 2.9,explode=(0.01,0.01,0.01,0.01,0.01,0.01,0.01),)

pie = plt.pie(doughnut[doughnut['cut']=='Premium']['count'],labeldistance = 1.1,autopct = '%0.01f%%',pctdistance=0.89,startangle = 90,
            wedgeprops = dict(width = 0.5),radius = 2.35,explode=(0.01,0.01,0.01,0.01,0.01,0.01,0.01),)

pie = plt.pie(doughnut[doughnut['cut']=='Very Good']['count'],labeldistance = 1.1,autopct = '%0.01f%%',pctdistance=0.85,startangle = 90,
            wedgeprops = dict(width = 0.5),radius = 1.8,explode=(0.01,0.01,0.01,0.01,0.01,0.01,0.01),)

plt.title('Relationship between cut and its colors', loc = 'center')
plt.axis('equal')
plt.legend(pie[0], labels = doughnut[doughnut['cut'] == 'Fair']['color'], loc = 'right',bbox_to_anchor = (1,0.7))
plt.show()
```

#### We learn
- In Fair cut, H and G are more
- In Ideal, E,F,G are more
- In premium, E,F,G,H are more
- Good,  E,F,G are more
- In Very Good, E and G are more

### 2- Parallel boxplot showing variation of price for each cut.

```python
fig = plt.figure(figsize=(20,10))  # is used to adjust the plot size
sns.boxplot(x = 'cut' , y ='price' , data = diamonds,palette = 'bright')
sns.set(font_scale=1.8)
plt.xlabel('Type of Cuts')
plt.ylabel('Price Range')
plt.title('Box plot of Cuts v/s Price Range')
plt.show()
```

#### The Following insights can be generated from the above graph.
- The price of the diamond increases according to its cuts, and as we can see that a diamond with premium cut is expensive followed by very good cut, and the rest. 
- Every cut has outliers , but the ideal cut (I.e normal cut) has the maximum number of outliers compared to the rest, which indicates that a diamond even with an ideal cut is sold on  larger scale.

### 3 Histogram for carat, depth,table and y.

```python
diamonds_1000 = diamonds.head(1000)
fig = plt.figure(figsize= (20,15) )
plt.subplot(2,2,1)    
plt.xlabel('carats')
plt.ylabel('No of carats')
plt.title('Histogram for Carat')
plt.hist(diamonds['carat'] , color = 'm')

plt.subplot(2,2,2)
plt.xlabel('Depth')
plt.ylabel('Count of depth')
plt.title('Histogram for Depth')
plt.hist(diamonds['depth'] , color = "cyan")
plt.tight_layout() 

plt.subplot(2,2,3)
plt.xlabel('Tables')
plt.ylabel('Count of Tables')
plt.title('Histogram for Table')
plt.hist(diamonds['table'] , color = "burlywood")

plt.subplot(2,2,4)
plt.xlabel('Y')
plt.ylabel('Count of Y')
plt.title('Histogram for Y')
plt.hist(diamonds['y'] , color = "chartreuse")
plt.tight_layout() 
```

#### The above Histogram for Crats gives us the following insights :
- The carats betweenn 0 -1 has the highest count and the carats between 2-3 is the least.
- The Carats between 0-1 is sold more than the rest.

#### The above Htogram for Depth shows :
- The Depth between 60 - 65 is the highest and that between 55-58 & 65-70 is less.
- More No of Diamonds is made with the depth between 60-65.

#### The above Histogram for Tables gives us the following insights :
- The count of table between 55-60 is highest and between 50-55 and 65-70 is less.

#### The above Histogram for Y shows :
- The count of y(width) is high between 0-5, which indicates more of diamonds are made with y(width) between 0-5

### 4- Joint plot showing variation of x and z (kind = Scatter)

```python
sns.jointplot(x = 'x' , y = 'z' , data = diamonds_1000, kind = 'scatter' , color =  'gold' ,
              marginal_kws=dict(bins=20),  size = 12)
plt.show()
```

#### The above joint plot shows :
- x(length) and z(depth) are proportional to each other, if x(length) is less z(depth) is also less, if X(length) is high, z(depth) is also high.
- More Number of diamonds are made with higher x(length) and higher z(depth) (I.e with x(length) between 5.5-6.5 and  z(depth) between 3-4)
- x(length)  & z(depth) both are bimodal.

### 8- Scatterplot between displacement and horsepower

```python
fig = plt.figure(figsize=(15,8))

#setting style
plt.style.use('seaborn-whitegrid')

#creating scatter plot

#for USA
plt.scatter(mpg[mpg['origin']=='usa']['displacement'],
           mpg[mpg['origin']=='usa']['horsepower'],
            c = 'coral',label = 'USA') #c stands for color
#for JAPAN
plt.scatter(mpg[mpg['origin']=='japan']['displacement'],
           mpg[mpg['origin']=='japan']['horsepower'],
            c = 'teal',label = 'JAPAN') #c stands for color
#for EUROPE
plt.scatter(mpg[mpg['origin']=='europe']['displacement'],
           mpg[mpg['origin']=='europe']['horsepower'],
            c = 'gold',label = 'EUROPE') #c stands for color


plt.legend()
plt.xlabel('Displacement')
plt.ylabel('Horsepower')
plt.title('Countrywise Scaterplot of DIST VS HPOWER')
plt.show()
```

### We see countrywise displacement vs horsepower chart for the data. Some key points 
- All of japan's cars lie between horsepower 40-130 wherein USA has cars ranging from horsepower 50 till 230.
- We can say traditionally USA has been making excellent cars in terms of horsepower & displacement almost doubling that of japan and europe , wherein europe & japan almost tag along with each other.

### 9- violin plot for acceleration with cylinders

```python
#figure size
plt.figure(figsize=(20,8))

#plotting violin plot
sns.violinplot(x= 'cylinders',y='acceleration',data = mpg,split = True, palette = 'bright')
plt.title('Cylinderwise Violinplot for Acceleration')
plt.show()
```

#### We learn that
- Interestingly We find out that cars with 4cylinders tend to have higher acceleration than the one having more than that.
- Anything more than 5 cylinders is an overkill in terms of acceleration for cars.
- 3 cylinder models have very less range.

### 10- Swarmplot for variation of horsepower for each model

```python
sns.swarmplot(x='model_year',y='horsepower',data = mpg,palette = 'Paired')
plt.title('Yearwise Swarmplot for Horsepower')
plt.show()
```

####  We learn 
- Over the years horsepower has seen drastic drop in terms of horsepower.
- The decline rate is almost close to 75%, One of the probable factor for this maybe the companies have started catering to customer needs. Hence in order to get more customers companies started producing majority of their models with higher degree of safety and lower cost thus substancially reducing the horsepower of a car.
