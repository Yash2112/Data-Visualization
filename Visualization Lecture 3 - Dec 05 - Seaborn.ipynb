```python
print(credit.columns)
print(credit['default payment next month'].value_counts()/credit.shape[0]*100)
credit.head(2)
```

```python
credit_features  =  credit.iloc[:,0:24]
credit_label     =  credit.iloc[:,24]
print(credit_features.shape)
print(credit_label.shape)
credit_features.columns
```

```python
credit_features.info()
```

```python

train_x, test_x, train_y, test_y = train_test_split(credit_features, credit_label, 
                                    test_size = 0.3,random_state = 100)
print("train_x", train_x.shape,"test_x", test_x.shape,"train_y", train_y.shape,"test_y", test_y.shape)
#test_y = pd.Series('test_y')
```

```python
model_rf = RandomForestClassifier(random_state=100, n_estimators=500)
model_rf.fit(train_x, train_y)

test_pred_rf = model_rf.predict(test_x)
df_pred_rf = pd.DataFrame({'actual' : test_y,
                       'predicted' : test_pred_rf})
df_pred_rf['pred_status'] = df_pred_rf['actual'] == df_pred_rf['predicted']
df_pred_rf['pred_status'].sum() / df_pred_rf.shape[0] * 100
```

```python
model_ab  = AdaBoostClassifier(random_state=100)
model_ab.fit(train_x,train_y)
test_pred_ab = model_ab.predict(test_x)
df_pred_ab = pd.DataFrame({'actual' : test_y,
                       'predicted' : test_pred_ab})
df_pred_ab['pred_status'] = df_pred_ab['actual'] == df_pred_ab['predicted']
df_pred_ab['pred_status'].sum() / df_pred_ab.shape[0] * 100
```
